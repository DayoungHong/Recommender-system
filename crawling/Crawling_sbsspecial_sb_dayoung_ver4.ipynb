```python
driver.get('http://programs.sbs.co.kr/culture/sbsspecial/vods/53591')
time.sleep(2)
body = driver.find_element_by_tag_name("body")
```

### 전체 목록 펼치기

```python
#최대 페이지는 프로그램 분량에 따라 변경가능. 약간 비효율적일 수도 있음.
#program-front-common-more-more-self 더 보기 버튼 누르기
num_of_pagedowns = 50
while num_of_pagedowns:
    body.send_keys(Keys.PAGE_DOWN)
    time.sleep(2)
    num_of_pagedowns -= 1
    try:
        driver.find_element_by_xpath("""//*[@id="program-front-common-more-more-self"]/button""").click()
    except:
        None
```

```python
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

#episodes = soup.find_all("span", "titprog_episode")
title = soup.find("em", "pheader_tit")
episodes = soup.find_all("em", "titprog_name")
dates = soup.find_all("span", "itxt_data")
links = soup.find_all("a", "ali_module_link")
```

```python
#단순 문자열 반복 뽑기 함수
#이 파일에서 데이터 추출에는 쓰지 않았음
'''def vod_info(n):
    for i in n:
        print (i.get_text())
'''
```

### 전체 에피소드 개수 세기

```python
vod_num=len(episodes)
vod_num
```

### column : vod_tv에 제작사 리스트 저장 (vod_tv)

```python
vod_tv=[]
tv="SBS"
for i in range(0, vod_num):
    vod_tv.append('SBS')
    
vod_tv
```

### column : title에 방송 프로그램 이름 리스트 저장 (vod_title)

```python
vod_title=[]

for tit in range(0, vod_num):
    title.get_text()
    vod_title.append(title.get_text())

    
vod_title
```

### column : epi에 각 회차 이름 리스트 저장 (vod_epi)

```python
vod_epi=[]
for epi in episodes:
    vod_epi.append(epi.get_text())
    #vod_name = pd.DataFrame(name)

print(vod_epi)


#vod_info(names)
```

### column : date에 방송일 리스트 저장 (vod_date)

```python
vod_date=[]
for date in dates:
    vod_date.append(date.get_text()[4:])

vod_date

#for date in dates:
#    print(date.get_text()[4:])
```

### column : link에 각 회차 링크 리스트 저장 (vod_link)

```python
vod_link=[]
for link in links:
    vod_link.append("http://programs.sbs.co.kr" + link['href'])

vod_link

#for a in links: 
#    print("http://programs.sbs.co.kr" + a['href'])
```

## 시놉시스 [synop, vod_synop]

### 미리보기 페이지로 이동

```python
driver.get('http://programs.sbs.co.kr/culture/sbsspecial/visualboards/53590')

time.sleep(2)
body = driver.find_element_by_tag_name("body")
```

### 전체 목록 보이기

```python
num_of_pagedowns = 50
while num_of_pagedowns:
    body.send_keys(Keys.PAGE_DOWN)
    time.sleep(2)
    num_of_pagedowns -= 1
    try:
        driver.find_element_by_xpath("""//*[@id="program-front-end-board-area"]/div/div/div[3]/div/div[1]/button""").click()
    except:
        None
```

### 각 회차의 링크 추출 후 링크를 vod_synoplink 리스트에 저장

```python
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

links = soup.find_all("a", "blml_module_link")

vod_synoplink = [ ]
for link in links:
    vod_synoplink.append(link['href'])                  
```

### 미리보기 링크로 회차 페이지 이동 후, 시놉시스 뽑아서 vod_synop 리스트 에 저장

```python
vod_synop=[]
for synoplink in vod_synoplink:  
    driver.get(synoplink)
    html=driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    contents=soup.find_all("div", "aba_inner")
    text = re.sub('<[^<]+?>', '', str(contents))
    text.replace("\xa0", "")
    vod_synop.append(text)
    
vod_synop
```

## 모든 column DataFrame에 저장

```python
vod_sbsspecial = {"tv":vod_tv, "title":vod_title, "episode":vod_epi, "date":vod_date, "link":vod_link, "synopsys":vod_synop}
df = pd.DataFrame.from_dict(vod_sbsspecial, orient='index')
df.transpose()
```
